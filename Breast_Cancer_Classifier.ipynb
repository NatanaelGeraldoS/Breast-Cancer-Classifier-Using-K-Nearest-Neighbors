```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
breast_cancer_data = load_breast_cancer()
```

```python
breast_cancer_data.data[0]
```

```python
breast_cancer_data.feature_names
```

```python
print(breast_cancer_data.target)
print(breast_cancer_data.target_names)
```

# Split Data Set

```python
training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size=0.2, random_state=100)
```

```python
print("Length of Training Data : ", len(training_data))
print("Length of Training Labels : ", len(training_labels))
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
```

# Fit DataSet

```python
classifier.fit(training_data, training_labels)
```

```python
print(classifier.score(validation_data, validation_labels))
```

# Iterate K Neighbours

```python
k_values = range(1, 101)
```

```python
accuracy_array = []
for i in k_values:
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(training_data, training_labels)
    accuracy_array.append(classifier.score(validation_data, validation_labels))
```

```python
print(accuracy_array)
```

```python
plt.clf()

plt.plot(k_values, accuracy_array)
plt.title('Breast Cancer Classifier Accuracy')
plt.xlabel('k')
plt.ylabel('Validation Accuracy')
plt.grid(True)

plt.show()
```
